```python
import matplotlib.pyplot as plt

# Create figure
fig, ax = plt.subplots(figsize=(9, 9))

# Draw Moon
moon = plt.Circle((0, 0), 1, color='lightgray', ec='black')
ax.add_patch(moon)

# Features: (name, x, y, description)
features = [
    ("Tycho", 0.0, -0.8,
     "Young impact crater with bright rays; named 1651 (Riccioli)"),

    ("Copernicus", -0.2, -0.3,
     "Large prominent crater; named 1651 (Riccioli)"),

    ("Kepler", -0.4, -0.1,
     "Small bright-ray crater; named 1651 (Riccioli)"),

    ("Aristarchus", -0.5, -0.2,
     "Brightest crater on Moon; named 1651 (Riccioli)"),

    ("Plato", -0.3, 0.6,
     "Dark-floored lava-filled crater; named 1651 (Riccioli)"),

    ("Mare Imbrium", -0.5, 0.1,
     "Large lava plain (basin); named 1651 (Riccioli)"),

    ("Mare Serenitatis", 0.3, 0.3,
     "Basaltic plain ('Sea of Serenity'); named 1651"),

    ("Mare Tranquillitatis", 0.6, 0.0,
     "Apollo 11 landing site; named 1651"),

    ("Mare Crisium", 0.8, 0.1,
     "Isolated circular basin; named 1651"),

    ("Montes Apenninus", 0.0, -0.1,
     "Mountain range on Imbrium edge; named 1651"),

    ("Montes Alpes", -0.3, 0.3,
     "Mountain range with Alpine Valley; named 1651"),
]

# Plot
for name, x, y, desc in features:
    ax.scatter(x, y, s=30)
    ax.text(
        x + 0.04, y + 0.04,
        f"{name}\n{desc}",
        fontsize=7,
        bbox=dict(facecolor='white', alpha=0.6, pad=1)
    )

# Formatting
ax.set_aspect('equal')
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.axis('off')

plt.title("Moon Map with Key Features (Annotated)", fontsize=14)
plt.show()
```
