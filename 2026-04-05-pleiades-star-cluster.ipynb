```python
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
import astropy.units as u

# Main Pleiades stars (Seven Sisters + Atlas & Pleione often included)
stars = {
    "Alcyone": ("03h47m29.1s", "+24d06m18s"),
    "Maia": ("03h45m49.6s", "+24d22m04s"),
    "Electra": ("03h44m52.5s", "+24d06m49s"),
    "Taygeta": ("03h45m12.5s", "+24d28m02s"),
    "Celaeno": ("03h44m48.2s", "+24d17m22s"),
    "Sterope": ("03h45m54.5s", "+23d57m24s"),
    "Merope": ("03h46m19.6s", "+23d56m54s"),
    "Atlas": ("03h49m09.7s", "+24d03m12s"),
    "Pleione": ("03h49m11.2s", "+24d08m12s"),
}

# Convert to SkyCoord
coords = SkyCoord(
    [v[0] for v in stars.values()],
    [v[1] for v in stars.values()],
    frame="icrs"
)

# Convert to degrees
ra = coords.ra.degree
dec = coords.dec.degree

# Center the plot around Alcyone (brightest star)
center_ra = ra[0]
center_dec = dec[0]

ra_offset = (ra - center_ra) * 60  # arcminutes
dec_offset = (dec - center_dec) * 60  # arcminutes

# Plot
plt.figure(figsize=(6, 6))
plt.scatter(ra_offset, dec_offset, s=100)

# Label stars
for i, name in enumerate(stars.keys()):
    plt.text(ra_offset[i] + 1, dec_offset[i] + 1, name, fontsize=9)

# Formatting
plt.gca().invert_xaxis()  # astronomical convention
plt.xlabel("RA offset (arcmin)")
plt.ylabel("Dec offset (arcmin)")
plt.title("Pleiades (Seven Sisters) - Main Stars")
plt.grid(True)

plt.show()
```
