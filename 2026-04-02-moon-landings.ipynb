```python
import matplotlib.pyplot as plt

# Create figure
fig, ax = plt.subplots(figsize=(9, 9))

# Draw Moon
moon = plt.Circle((0, 0), 1, color='lightgray', ec='black')
ax.add_patch(moon)

# Landing sites: (name, x, y, description)
sites = [
    # Apollo missions (human)
    ("Apollo 11", 0.6, 0.0, "First human landing (1969)"),
    ("Apollo 12", 0.3, -0.2, "Precision landing near Surveyor 3 (1969)"),
    ("Apollo 14", -0.2, -0.4, "Fra Mauro formation (1971)"),
    ("Apollo 15", 0.1, 0.2, "Hadley Rille, first rover (1971)"),
    ("Apollo 16", 0.4, 0.4, "Highlands mission (1972)"),
    ("Apollo 17", 0.5, 0.2, "Last human mission (1972)"),

    # Soviet Luna missions (robotic)
    ("Luna 2", 0.2, 0.6, "First impact on Moon (1959)"),
    ("Luna 9", -0.4, 0.1, "First soft landing (1966)"),
    ("Luna 16", 0.0, -0.3, "Sample return (1970)"),
    ("Luna 17", -0.5, 0.3, "Lunokhod 1 rover (1970)"),
    ("Luna 21", -0.6, 0.4, "Lunokhod 2 rover (1973)"),

    # China missions
    ("Chang'e 3", -0.3, 0.2, "Yutu rover (2013)"),
    ("Chang'e 4", -0.8, -0.1, "First far-side landing (2019)"),
    ("Chang'e 5", 0.2, -0.1, "Sample return (2020)"),

    # India
    ("Chandrayaan-3", 0.1, -0.8, "South pole landing (2023)"),

    # Others
    ("Surveyor 1", -0.2, 0.0, "NASA soft landing (1966)"),
]

# Plot points and labels
for name, x, y, desc in sites:
    ax.scatter(x, y, s=35)
    ax.text(
        x + 0.04, y + 0.04,
        f"{name}\n{desc}",
        fontsize=7,
        bbox=dict(facecolor='white', alpha=0.6, pad=1)
    )

# Formatting
ax.set_aspect('equal')
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.axis('off')

plt.title("Moon Landing Sites", fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(9, 9))

# Draw Moon
moon = plt.Circle((0, 0), 1, color='lightgray', ec='black')
ax.add_patch(moon)

# Divide near side / far side (vertical line)
ax.plot([0, 0], [-1, 1], linestyle='--')

# Optional shading (far side slightly darker)
far_side = plt.Circle((0, 0), 1, color='gray', alpha=0.3)
ax.add_patch(far_side)
ax.add_patch(plt.Rectangle((0, -1), 1, 2, color='gray', alpha=0.3))

# Landing sites
sites = [
    ("Apollo 11", 0.6, 0.0, "1969"),
    ("Apollo 12", 0.3, -0.2, "1969"),
    ("Apollo 14", -0.2, -0.4, "1971"),
    ("Apollo 15", 0.1, 0.2, "1971"),
    ("Apollo 16", 0.4, 0.4, "1972"),
    ("Apollo 17", 0.5, 0.2, "1972"),
    ("Luna 9", -0.4, 0.1, "1966"),
    ("Chang'e 4", -0.8, -0.1, "Far side (2019)"),
    ("Chandrayaan-3", 0.1, -0.8, "2023"),
]

# Plot points + labels
for name, x, y, desc in sites:
    ax.scatter(x, y, s=35)
    ax.text(
        x + 0.04, y + 0.04,
        f"{name}\n{desc}",
        fontsize=7,
        bbox=dict(facecolor='white', alpha=0.6, pad=1)
    )

# Labels for sides
ax.text(-0.5, 1.05, "Near Side", ha='center', fontsize=10)
ax.text(0.5, 1.05, "Far Side", ha='center', fontsize=10)

# Formatting
ax.set_aspect('equal')
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.axis('off')

plt.title("Moon Landing Sites (Near vs Far Side)")
plt.show()
```
