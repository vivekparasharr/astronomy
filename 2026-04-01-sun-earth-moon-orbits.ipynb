```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------
# REAL VALUES (meters)
# ----------------------------
R_sun = 696340e3       # Sun radius
R_earth = 6371e3       # Earth radius
R_moon = 1737.4e3      # Moon radius

AU = 149597870700      # Earth-Sun distance
earth_moon_dist = 384400e3  # Earth-Moon distance

# ----------------------------
# SCALE SETTINGS
# ----------------------------
DIST_SCALE = 1 / AU     # 1 AU -> 1 unit
SIZE_SCALE = 2000       # Increase to make planets visible

# ----------------------------
# SCALED VALUES
# ----------------------------
sun_radius = R_sun * DIST_SCALE * SIZE_SCALE
earth_radius = R_earth * DIST_SCALE * SIZE_SCALE
moon_radius = R_moon * DIST_SCALE * SIZE_SCALE

earth_orbit_radius = AU * DIST_SCALE
moon_orbit_radius = earth_moon_dist * DIST_SCALE

# ----------------------------
# POSITIONS
# ----------------------------
sun_pos = np.array([0, 0])
earth_pos = np.array([earth_orbit_radius, 0])
moon_pos = earth_pos + np.array([moon_orbit_radius, 0])

# ----------------------------
# PLOT
# ----------------------------
fig, ax = plt.subplots(figsize=(8, 8))

# Sun
ax.add_patch(plt.Circle(sun_pos, sun_radius, color='yellow', label='Sun'))

# Earth orbit
ax.add_patch(plt.Circle(sun_pos, earth_orbit_radius, fill=False, linestyle='--'))

# Earth
ax.add_patch(plt.Circle(earth_pos, earth_radius, color='blue', label='Earth'))

# Moon orbit
ax.add_patch(plt.Circle(earth_pos, moon_orbit_radius, fill=False, linestyle=':'))

# Moon
ax.add_patch(plt.Circle(moon_pos, moon_radius, color='gray', label='Moon'))

# ----------------------------
# FORMAT
# ----------------------------
ax.set_aspect('equal')
ax.set_title("Sun–Earth–Moon System (Scaled, No Astropy)")

ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)

ax.legend()
plt.grid(True)
plt.show()
```
