```python
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
import astropy.units as u
import numpy as np

# -------------------------------
# Define 10 interesting MW objects
# -------------------------------
objects = [
    {"name": "Solar System", "l": 0, "b": 0, "distance": 0},
    {"name": "Sagittarius A*", "l": 0, "b": 0, "distance": 26000},
    {"name": "Betelgeuse", "l": 200, "b": -9, "distance": 642},
    {"name": "Vega", "l": 67, "b": 19, "distance": 25},
    {"name": "Pleiades", "l": 166, "b": -24, "distance": 444},
    {"name": "Orion Nebula", "l": 209, "b": -19, "distance": 1344},
    {"name": "Crab Nebula", "l": 184, "b": -6, "distance": 6500},
    {"name": "Cygnus X-1", "l": 71, "b": 3, "distance": 7200},
    {"name": "47 Tucanae", "l": 305, "b": -45, "distance": 16000},
    {"name": "Large Magellanic Cloud", "l": 280, "b": -33, "distance": 163000},
]

# -------------------------------
# Convert to Cartesian coordinates
# -------------------------------
x_vals = []
y_vals = []
labels = []

for obj in objects:
    coord = SkyCoord(
        l=obj["l"] * u.degree,
        b=obj["b"] * u.degree,
        distance=obj["distance"] * u.lightyear,
        frame="galactic"
    )
    
    # Convert to cartesian
    cart = coord.cartesian
    x_vals.append(cart.x.value)
    y_vals.append(cart.y.value)
    labels.append(obj["name"])

# -------------------------------
# Plot
# -------------------------------
plt.figure(figsize=(10, 10))

plt.scatter(x_vals, y_vals)

# Annotate points
for i, label in enumerate(labels):
    plt.text(x_vals[i], y_vals[i], label, fontsize=9)

# Styling
plt.axhline(0)
plt.axvline(0)
plt.title("Top 10 Interesting Objects in the Milky Way (Approximate Map)")
plt.xlabel("X (light-years)")
plt.ylabel("Y (light-years)")
plt.grid(True)

plt.show()
```
